```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
doctorado = pd.read_csv('Doctorado.csv')
doctorado
```

```python
tiene_doct = pd.read_csv('Tiene_Doctorado.csv')
tiene_doct
```

```python
doctorado_tie = pd.DataFrame(doctorado.Confirmacion.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["Ninguno","Tiene"]
doctorado_plot = ax.bar(doctorado_tie.index, doctorado_tie.Confirmacion, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(doctorado_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CON DOCTORADO')
ax.legend(title='Doctorado')
```

```python
infodocto = pd.DataFrame(tiene_doct.Doctorado.value_counts())
fig, ax = plt.subplots()
infodocto_bar = ax.barh(infodocto.index, infodocto.Doctr, align='center', height=0.7)
ax.bar_label(infodocto_bar, padding=1)
plt.xlabel("Cantidad")
plt.title("AREA DE DOCTORADO")
```

```python
agru_sex = pd.DataFrame(tiene_doct.groupby(['SEXO','Confirmacion']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON DOCTORADO: DOCTORADO VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
