```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
diplomad1 = pd.read_csv('Diplomado_1.csv')
diplomad1
```

```python
#Personas encontradas que hicieron diplomado
diplomado1 = pd.DataFrame(diplomad1.Confirmacion.value_counts())
diplomado1
```

```python
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["Ninguno","Tiene"]
diplomado1_plot = ax.bar(diplomado1.index, diplomado1.Confirmacion, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(diplomado1_plot, padding=1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CON 1 DIPLOMADO')
ax.legend(title='Diplomado 1')

```
