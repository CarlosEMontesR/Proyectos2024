```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
especializacion2 = pd.read_csv('Especializacion2.csv')
especializacion2
```

```python
tiene_espe2 = pd.read_csv('Tiene_Especializacion2.csv')
tiene_espe2
```

```python
espe1_tie = pd.DataFrame(especializacion2.Confirmacion.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["Ninguno","Tiene"]
especializacion2_plot = ax.bar(espe1_tie.index, espe1_tie.Confirmacion, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(especializacion2_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CON 2 ESPECIALIZACIÓN')
ax.legend(title='Especialización 2')
```

```python
infoespe2 = pd.DataFrame(tiene_espe2.Especializacion_2.value_counts())
fig, ax = plt.subplots()
infoespe2_bar = ax.barh(infoespe2.index, infoespe2.Especializacion_2, align='center', height=0.7)
ax.bar_label(infoespe2_bar, padding=1)
plt.xlabel("Cantidad")
plt.title("AREA DE 2da ESPECIALIZACIÓN")
```

```python
agru_sex = pd.DataFrame(tiene_espe2.groupby(['SEXO','Confirmacion']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON 2 ESPECIALIZACIONES: ESPECIALIZACIONES VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```

```python
agru_sex = pd.DataFrame(tiene_espe2.groupby(['SEXO','Anio']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS ESPECIALIZADOS: GENERO vs AÑO")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
