```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
linkedin = pd.read_csv("LinkedInV1.csv")
```

```python
buscadas = pd.DataFrame(linkedin.Año.value_counts())
fig, ax = plt.subplots()
bar_colors=['red','blue','green','yellow','orange','purple','pink','brown','orange','green']
bar_labels=['tab:red','tab:blue','tab:green','tab:yellow','tab:orange','tab:purple','tab:pink','tab:brown','tab:orange','tab:green']
graficabuscadas = ax.bar(buscadas.index,buscadas.Año, width=0.4, align='center', label=bar_labels, color=bar_colors)
ax.bar_label(graficabuscadas, padding=3)
plt.title("EGRESADOS BUSCADOS POR AÑO")
ax.set_ylabel('Cantidad')
ax.set_xlabel('Año')
plt.grid()
```

```python
encontradas = pd.DataFrame(linkedin.LinkedIn.value_counts())
fig, ax = plt.subplots()
bar_colors=['blue','green']
bar_labels=['tab:red','tab:blue']
myplot2 = ax.bar(encontradas.index,encontradas.LinkedIn, width=0.8, align='center', label=bar_labels, color=bar_colors)
ax.bar_label(myplot2, padding=3)
plt.title("EGRESADOS BUSCADOS")
ax.set_ylabel('Cantidad')
ax.set_xlabel('Estado')
```

```python
fig, ax = plt.subplots()
bar_colors=['blue','green']
bar_labels=['tab:red','tab:blue']
myplot = ax.bar(encontradas.index,encontradas.SEXO, width=0.8, align='center', label=bar_labels, color=bar_colors)
ax.bar_label(myplot, padding=3)
plt.title("Personas buscadas vs Genero")
ax.set_ylabel('Cantidad')
ax.set_xlabel('Genero')
```

```python
datosagrupados = linkedin.groupby(['Año','SEXO']).size().unstack()
datosagrupados.plot(kind="bar", stacked=False)
plt.title('Egresados por genero y año de graduado')
plt.xlabel('Año de graduado')
plt.ylabel('Cantidad de egresados')
plt.grid()
plt.show()
```

```python
egresados_encontrados = pd.read_csv('Encontrados.csv')
egresados_encontrados
```

```python
egre_encon = pd.DataFrame(egresados_encontrados.SEXO.value_counts())
fig, ax = plt.subplots()
bar_colors=['blue','green']
bar_labels=['tab:red','tab:blue']
egre_encon_plot = ax.bar(egre_encon.index,egre_encon.SEXO, width=0.8, align='center', label=bar_labels, color=bar_colors)
ax.bar_label(egre_encon_plot, padding=3)
plt.title("EGRESADOS ENCONTRADOS")
ax.set_ylabel('Cantidad')
ax.set_xlabel('GENERO')
```

```python
genero_año_new = egresados_encontrados.groupby(['Año','SEXO']).size().unstack()
genero_año_new.plot(kind="bar", stacked=False)
plt.title('EGRESADOS ENCONTRADOS: GENEROS VS AÑOS')
plt.xlabel('Año de graduado')
plt.ylabel('Cantidad de egresados')
plt.grid()
plt.show()
```
