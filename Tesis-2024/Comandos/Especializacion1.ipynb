```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
especializacion1 = pd.read_csv("Especializacion1.csv")
especializacion1
```

```python
tiene_esp_1 = pd.read_csv("Tiene_Especializacion1.csv")
tiene_esp_1
```

```python
espe1_tie = pd.DataFrame(especializacion1.Confirmacion.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["Ninguno","Tiene"]
diplomado2_plot = ax.bar(espe1_tie.index, espe1_tie.Confirmacion, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(diplomado2_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CON 1 ESPECIALIZACIÓN')
ax.legend(title='Especialización 1')
```

```python
infoespe1 = pd.DataFrame(tiene_esp_1.Especializacion1.value_counts())
fig, ax = plt.subplots()
infoespe1_bar = ax.barh(infoespe1.index, infoespe1.Especializacion1, align='center', height=0.7)
ax.bar_label(infoespe1_bar, padding=1)
plt.xlabel("Cantidad")
plt.title("AREA DE 2da ESPECIALIZACIÓN")
```

```python
agru_sex = pd.DataFrame(tiene_esp_1.groupby(['SEXO','Confirmacion']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.title("CANTIDAD DE EGRESADOS CON 2da ESPECIALIZACION POR GENERO")
plt.grid()
```

```python
agru_sex = pd.DataFrame(tiene_esp_1.groupby(['SEXO','Anio']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS ESPECIALIZADOS: GENERO vs AÑO")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
