```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
administrativo1 = pd.read_csv('Administrativo_1.csv')
administrativo1
```

```python
tiene_administrativo1 = pd.read_csv('TrabajoAdministrativo1.csv')
tiene_administrativo1
```

```python
administrativo1_tie = pd.DataFrame(administrativo1.Trabajo.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
administrativo1_plot = ax.bar(administrativo1_tie.index, administrativo1_tie.Trabajo, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(administrativo1_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS EN PUESTOS ADMINISTRATIVOS')
ax.legend(title='Docente')
```

```python
infoadmin1 = pd.DataFrame(tiene_administrativo1.Administrativo1.value_counts())
fig, ax = plt.subplots()
infoadmin1_bar = ax.barh(infoadmin1.index, infoadmin1.Administrativo1, align='center', height=0.7)
ax.bar_label(infoadmin1_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EGRESADOS COMO DOCENTES EN INSTITUCIONES")
```

```python
agru_sex = pd.DataFrame(tiene_administrativo1.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON ADMINISTRATIVO: DOCENCIA VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
