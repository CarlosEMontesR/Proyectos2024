```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
docencia3 = pd.read_csv('Docencia_3.csv')
docencia3
```

```python
tiene_docencia3 = pd.read_csv('TrabajoDocencia3.csv')
tiene_docencia3
```

```python
docencia3_tie = pd.DataFrame(docencia3.Docente.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
docencia3_plot = ax.bar(docencia3_tie.index, docencia3_tie.Docente, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(docencia3_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS QUE TRABAJARON EN 3 O MÁS INSTITUCIONES COMO DOCENTES')
ax.legend(title='Docente')
```

```python
infodoce3 = pd.DataFrame(tiene_docencia3.Docencia_3.value_counts())
fig, ax = plt.subplots()
infodocencia3_bar = ax.barh(infodoce3.index, infodoce3.Docencia_3, align='center', height=0.7)
ax.bar_label(infodocencia3_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EGRESADOS COMO DOCENTES EN INSTITUCIONES")
```

```python
agru_sex = pd.DataFrame(tiene_docencia3.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON DOCENCIA: DOCENCIA VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
