```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
director2 = pd.read_csv('Director_2.csv')
director2
```

```python
tiene_director2 = pd.read_csv('TrabajoDirector2.csv')
tiene_director2
```

```python
director2_tie = pd.DataFrame(director2.Trabajo.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
director2_plot = ax.bar(director2_tie.index, director2_tie.Trabajo, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(director2_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS DIRECTORES')
ax.legend(title='Director')
```

```python
infodirector2 = pd.DataFrame(tiene_director2.Director_2.value_counts())
fig, ax = plt.subplots()
infodirector2_bar = ax.barh(infodirector2.index, infodirector2.Director_2, align='center', height=0.7)
ax.bar_label(infodirector2_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EMPRESAS:EGRESADOS COMO DIRECTORES")
```

```python
agru_sex = pd.DataFrame(tiene_director2.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS DIRECTORES: EGRESADOS VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
