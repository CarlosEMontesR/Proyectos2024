```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
diplomado_2 = pd.read_csv("Diplomado_2.csv")
diplomado_2
```

```python
dip_2_tie = pd.DataFrame(diplomado_2.Confirmacion.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["Ninguno","Tiene"]
diplomado2_plot = ax.bar(dip_2_tie.index, dip_2_tie.Confirmacion, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(diplomado2_plot, padding=1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CON 2 DIPLOMADOS')
ax.legend(title='Diplomado 2')
```

```python
tiene_dip2 = pd.read_csv("Tiene_2Diplomado.csv")
tiene_dip2
```

```python
infodip2 = pd.DataFrame(tiene_dip2.Diplomado_2.value_counts())
fig, ax = plt.subplots()
infodip2_bar = ax.barh(infodip2.index, infodip2.Diplomado_2, align='center')
ax.set_title('TIPO DE SEGUNDO DIPLOMADO')
ax.set_xlabel('Cantidad de egresados con el mismo 2do diplomado')
ax.bar_label(infodip2_bar, padding=1)
```

```python
anio_egre = pd.DataFrame(tiene_dip2.A__o.value_counts())
#group_anio = pd.DataFrame(tiene_dip.groupby("A__o"))
#group_anio
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["Ninguno","Tiene"]
anio_egre_plot = ax.bar(anio_egre.index, anio_egre.A__o, width=0.4, align='center')
ax.bar_label(anio_egre_plot, padding=1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CON 2 DIPLOMADO vs AÑO DE EGRESADO')
plt.grid()
```

```python
agru_sex = pd.DataFrame(tiene_dip2.groupby(['SEXO','A__o']).size().unstack())
pltagru = agru_sex.plot(kind="bar", stacked=False)
plt.title('EGRESADOS CON 2DO DIPLOMADO POR GENERO')
plt.xlabel('Género')
plt.ylabel('Cantidad de egresados')
plt.grid()
plt.show()
```
