```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
docencia1 = pd.read_csv('Docencia_1.csv')
docencia1
```

```python
tiene_docencia1 = pd.read_csv('TrabajoenDocencia1.csv')
tiene_docencia1
```

```python
docencia_tie = pd.DataFrame(docencia1.Docente.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
doctorado_plot = ax.bar(docencia_tie.index, docencia_tie.Docente, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(doctorado_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS QUE TRABAJARON EN DOCENCIA')
ax.legend(title='Docente')
```

```python
infodoce1 = pd.DataFrame(tiene_docencia1.Docencia_1.value_counts())
fig, ax = plt.subplots()
infodocto_bar = ax.barh(infodoce1.index, infodoce1.Docencia_1, align='center', height=0.7)
ax.bar_label(infodocto_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EGRESADOS COMO DOCENTES EN INSTITUCIONES")
```

```python
agru_sex = pd.DataFrame(tiene_docencia1.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON DOCENCIA: DOCENCIA VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
