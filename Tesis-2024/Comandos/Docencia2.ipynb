```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
docencia2 = pd.read_csv('Docencia_2.csv')
docencia2
```

```python
tiene_docencia2 = pd.read_csv('TrabajoDocencia2.csv')
tiene_docencia2
```

```python
docencia2_tie = pd.DataFrame(docencia2.Docente.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
docencia2_plot = ax.bar(docencia2_tie.index, docencia2_tie.Docente, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(docencia2_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS QUE TRABAJARON EN 2 INSTITUCIONES COMO DOCENTES')
ax.legend(title='Docente')
```

```python
infodoce2 = pd.DataFrame(tiene_docencia2.Docencia_2.value_counts())
fig, ax = plt.subplots()
infodocencia2_bar = ax.barh(infodoce2.index, infodoce2.Docencia_2, align='center', height=0.7)
ax.bar_label(infodocencia2_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EGRESADOS COMO DOCENTES EN INSTITUCIONES")
```

```python
agru_sex = pd.DataFrame(tiene_docencia2.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON DOCENCIA: DOCENCIA VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
