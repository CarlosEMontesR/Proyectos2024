```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
consultor1 = pd.read_csv('Consultor_1.csv')
consultor1
```

```python
tiene_consultor1 = pd.read_csv('TrabajoConsultor1.csv')
tiene_consultor1
```

```python
consultor1_tie = pd.DataFrame(consultor1.Trabajo.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
consultor1_plot = ax.bar(consultor1_tie.index, consultor1_tie.Trabajo, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(consultor1_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CONSULTORES')
ax.legend(title='Consultor')
```

```python
infoconsultor1 = pd.DataFrame(tiene_consultor1.Consultor_1.value_counts())
fig, ax = plt.subplots()
infoconsultor1_bar = ax.barh(infoconsultor1.index, infoconsultor1.Consultor_1, align='center', height=0.7)
ax.bar_label(infoconsultor1_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EMPRESAS:EGRESADOS COMO CONSULTORES")
```

```python
agru_sex = pd.DataFrame(tiene_consultor1.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CONSULTORES: EGRESADOS VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
