```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
consultor2 = pd.read_csv('Consultor_2.csv')
consultor2
```

```python
tiene_consultor2 = pd.read_csv('TrabajoConsultor2.csv')
tiene_consultor2
```

```python
consultor2_tie = pd.DataFrame(consultor2.Trabajo.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
consultor2_plot = ax.bar(consultor2_tie.index, consultor2_tie.Trabajo, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(consultor2_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CONSULTORES')
ax.legend(title='Consultor')
```

```python
infoconsultor2 = pd.DataFrame(tiene_consultor2.Consultor_2.value_counts())
fig, ax = plt.subplots()
infoconsultor2_bar = ax.barh(infoconsultor2.index, infoconsultor2.Consultor_2, align='center', height=0.7)
ax.bar_label(infoconsultor2_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EMPRESAS:EGRESADOS COMO CONSULTORES")
```

```python
agru_sex = pd.DataFrame(tiene_consultor2.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CONSULTORES: EGRESADOS VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
