```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
coordinador1 = pd.read_csv('Coordinador_1.csv')
coordinador1
```

```python
tiene_coordinador1 = pd.read_csv('TrabajoCoordinador1.csv')
tiene_coordinador1
```

```python
coordinador1_tie = pd.DataFrame(coordinador1.Trabajo.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
coordinador1_plot = ax.bar(coordinador1_tie.index, coordinador1_tie.Trabajo, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(coordinador1_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS COORDINADORES')
ax.legend(title='Consultor')
```

```python
infocoordinador1 = pd.DataFrame(tiene_coordinador1.Coordinador_1.value_counts())
fig, ax = plt.subplots()
infocoordinador1_bar = ax.barh(infocoordinador1.index, infocoordinador1.Coordinador_1, align='center', height=0.7)
ax.bar_label(infocoordinador1_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EMPRESAS:EGRESADOS COMO COORDINADORES")
```

```python
agru_sex = pd.DataFrame(tiene_coordinador1.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS COORDINADORES: EGRESADOS VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
