```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
maestria2 = pd.read_csv('Maestria2.csv')
tiene_maestria2 = pd.read_csv('Tiene_Maestria2.csv')
```

```python
maestria2
```

```python
tiene_maestria2
```

```python
maes2_tie = pd.DataFrame(maestria2.Confirmacion.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["Ninguno","Tiene"]
maestria1_plot = ax.bar(maes2_tie.index, maes2_tie.Confirmacion, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(maestria1_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CON 2 MAESTRÍA')
ax.legend(title='Maestría 2')
```

```python
infomaes2 = pd.DataFrame(tiene_maestria2.Maestria_2.value_counts())
fig, ax = plt.subplots()
infomaes2_bar = ax.barh(infomaes2.index, infomaes2.Maestria_2, align='center', height=0.7)
ax.bar_label(infomaes2_bar, padding=1)
plt.xlabel("Cantidad")
plt.title("AREA DE 2da MAESTRíA")
```

```python
agru_sex = pd.DataFrame(tiene_maestria2.groupby(['SEXO','Confirmacion']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON 2 MAESTRÍA: MAESTRIA VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```

```python
agru_sex = pd.DataFrame(tiene_maestria2.groupby(['SEXO','Ano']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON MAESTRÍA: GENERO vs AÑO")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
