```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
administrativo_2 = pd.read_csv('Administrativo_2.csv')
administrativo_2
```

```python
tiene_administrativo2 = pd.read_csv('TrabajoAdministrativo2.csv')
tiene_administrativo2
```

```python
administrativo2_tie = pd.DataFrame(administrativo_2.Trabajo.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
administrativo2_plot = ax.bar(administrativo2_tie.index, administrativo2_tie.Trabajo, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(administrativo2_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS EN 2 PUESTOS ADMINISTRATIVOS')
ax.legend(title='Administrativos')
```

```python
infoadmin2 = pd.DataFrame(tiene_administrativo2.Administrativo_2.value_counts())
fig, ax = plt.subplots()
infoadmin2_bar = ax.barh(infoadmin2.index, infoadmin2.Administrativo_2, align='center', height=0.7)
ax.bar_label(infoadmin2_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EMPRESAS:EGRESADOS CON 2 AREAS ADMINISTRATIVAS")
```

```python
agru_sex = pd.DataFrame(tiene_administrativo2.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON ADMINISTRATIVO: EGRESADOS VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
