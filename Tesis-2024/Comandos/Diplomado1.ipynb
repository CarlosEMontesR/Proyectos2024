```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
diplomad1 = pd.read_csv('Diplomado_1.csv')
diplomad1
```

```python
#Personas encontradas que hicieron diplomado
diplomado1 = pd.DataFrame(diplomad1.Confirmacion.value_counts())
diplomado1
```

```python
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["Ninguno","Tiene"]
diplomado1_plot = ax.bar(diplomado1.index, diplomado1.Confirmacion, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(diplomado1_plot, padding=1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CON 1 DIPLOMADO')
ax.legend(title='Diplomado 1')

```

```python
tiene_dip = pd.read_csv('Tiene_Diplomado.csv')
tiene_dip
```

```python
infodip = pd.DataFrame(tiene_dip.Diplomado_1.value_counts())
fig, ax = plt.subplots()
infodip_bar = ax.barh(infodip.index, infodip.Diplomado_1, align='center')
ax.set_title('TIPO DE PRIMER DIPLOMADO')
ax.set_xlabel('Cantidad de egresados con el mismo 1er diplomado')
ax.bar_label(infodip_bar, padding=1)
```

```python
anio_egre = pd.DataFrame(tiene_dip.A__o.value_counts())
#group_anio = pd.DataFrame(tiene_dip.groupby("A__o"))
#group_anio
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["Ninguno","Tiene"]
anio_egre_plot = ax.bar(anio_egre.index, anio_egre.A__o, width=0.4, align='center')
ax.bar_label(anio_egre_plot, padding=1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CON 1 DIPLOMADO vs AÑO DE EGRESADO')
plt.grid()

```
