```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
coordinador3 = pd.read_csv('Coordinador_3.csv')
coordinador3
```

```python
tiene_coordinador3 = pd.read_csv('TrabajoCoordinador3.csv')
tiene_coordinador3
```

```python
coordinador3_tie = pd.DataFrame(coordinador3.Trabajo.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
coordinador3_plot = ax.bar(coordinador3_tie.index, coordinador3_tie.Trabajo, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(coordinador3_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS COORDINADORES')
ax.legend(title='Coordinador')
```

```python
infocoordinador3 = pd.DataFrame(tiene_coordinador3.Coordinador_3.value_counts())
fig, ax = plt.subplots()
infocoordinador3_bar = ax.barh(infocoordinador3.index, infocoordinador3.Coordinador_3, align='center', height=0.7)
ax.bar_label(infocoordinador3_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EMPRESAS:EGRESADOS COMO COORDINADORES")
```

```python
agru_sex = pd.DataFrame(tiene_coordinador3.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS COORDINADORES: EGRESADOS VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
