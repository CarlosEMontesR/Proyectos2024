```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
administrativo_3 = pd.read_csv('Administrativo_3.csv')
administrativo_3
```

```python
tiene_administrativo3 = pd.read_csv('TrabajoAdministrativo3.csv')
tiene_administrativo3
```

```python
administrativo3_tie = pd.DataFrame(administrativo_3.Trabajo.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
administrativo3_plot = ax.bar(administrativo3_tie.index, administrativo3_tie.Trabajo, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(administrativo3_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS EN +3 PUESTOS ADMINISTRATIVOS')
ax.legend(title='Administrativos')
```

```python
infoadmin3 = pd.DataFrame(tiene_administrativo3.Administrativo_3.value_counts())
fig, ax = plt.subplots()
infoadmin3_bar = ax.barh(infoadmin3.index, infoadmin3.Administrativo_3, align='center', height=0.7)
ax.bar_label(infoadmin3_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EMPRESAS:EGRESADOS CON +3 AREAS ADMINISTRATIVAS")
```

```python
agru_sex = pd.DataFrame(tiene_administrativo3.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON ADMINISTRATIVO: EGRESADOS VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
