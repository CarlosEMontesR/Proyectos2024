```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
director1 = pd.read_csv('Director_1.csv')
director1
```

```python
tiene_director1 = pd.read_csv('TrabajoDirector1.csv')
tiene_director1
```

```python
director1_tie = pd.DataFrame(director1.Trabajo.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
director1_plot = ax.bar(director1_tie.index, director1_tie.Trabajo, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(director1_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS EN +3 PUESTOS ADMINISTRATIVOS')
ax.legend(title='Administrativos')
```

```python
infodirector1 = pd.DataFrame(tiene_director1.Director_1.value_counts())
fig, ax = plt.subplots()
infodirector1_bar = ax.barh(infodirector1.index, infodirector1.Director_1, align='center', height=0.7)
ax.bar_label(infodirector1_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EMPRESAS:EGRESADOS COMO DIRECTORES")
```

```python
agru_sex = pd.DataFrame(tiene_director1.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS DIRECTORES: EGRESADOS VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
