```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
```

```python
director3 = pd.read_csv('Director_3.csv')
director3
```

```python
tiene_director3 = pd.read_csv('TrabajoDirector3.csv')
tiene_director3
```

```python
director3_tie = pd.DataFrame(director3.Trabajo.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
director3_plot = ax.bar(director3_tie.index, director3_tie.Trabajo, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(director3_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS DIRECTORES')
ax.legend(title='Director')
```

```python
infodirector3 = pd.DataFrame(tiene_director3.Director_3.value_counts())
fig, ax = plt.subplots()
infodirector3_bar = ax.barh(infodirector3.index, infodirector3.Director_3, align='center', height=0.7)
ax.bar_label(infodirector3_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EMPRESAS:EGRESADOS COMO DIRECTORES")
```

```python
agru_sex = pd.DataFrame(tiene_director3.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS DIRECTORES: EGRESADOS VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
