```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
maestria1 = pd.read_csv('Maestria1.csv')
tiene_maestria1 = pd.read_csv('Tiene_Maestria1.csv')
```

```python
maestria1
```

```python
tiene_maestria1
```

```python
maes1_tie = pd.DataFrame(maestria1.Confirmacion.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["Ninguno","Tiene"]
maestria1_plot = ax.bar(maes1_tie.index, maes1_tie.Confirmacion, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(maestria1_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS CON 1 MAESTRÍA')
ax.legend(title='Maestría 1')
```

```python
infomaes1 = pd.DataFrame(tiene_maestria1.Maestria_1.value_counts())
fig, ax = plt.subplots()
infomaes1_bar = ax.barh(infomaes1.index, infomaes1.Maestria_1, align='center', height=0.7)
ax.bar_label(infomaes1_bar, padding=1)
plt.xlabel("Cantidad")
plt.title("AREA DE 1ra MAESTRíA")
```

```python
agru_sex = pd.DataFrame(tiene_maestria1.groupby(['SEXO','Confirmacion']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON 1 MAESTRÍA: MAESTRIA VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```

```python
agru_sex = pd.DataFrame(tiene_maestria1.groupby(['SEXO','Ano']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS CON MAESTRÍA: GENERO vs AÑO")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
