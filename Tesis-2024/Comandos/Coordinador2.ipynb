```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
coordinador2 = pd.read_csv('Coordinador_2.csv')
coordinador2
```

```python
tiene_coordinador2 = pd.read_csv('TrabajoCoordinador2.csv')
tiene_coordinador2
```

```python
coordinador2_tie = pd.DataFrame(coordinador2.Trabajo.value_counts())
fig, ax = plt.subplots()
color_bar = ["red","blue"]
color_label = ["No","Si"]
coordinador2_plot = ax.bar(coordinador2_tie.index, coordinador2_tie.Trabajo, width=0.4, align='center', color=color_bar, label=color_label)
ax.bar_label(coordinador2_plot, padding=-1)
ax.set_ylabel('Cantidad de Egresados')
ax.set_title('EGRESADOS COORDINADORES')
ax.legend(title='Coordinador')
```

```python
infocoordinador2 = pd.DataFrame(tiene_coordinador2.Coordinador_2.value_counts())
fig, ax = plt.subplots()
infocoordinador2_bar = ax.barh(infocoordinador2.index, infocoordinador2.Coordinador_2, align='center', height=0.7)
ax.bar_label(infocoordinador2_bar, padding=1)
plt.xlabel("Cantidad")
plt.title(" EMPRESAS:EGRESADOS COMO COORDINADORES")
```

```python
agru_sex = pd.DataFrame(tiene_coordinador2.groupby(['SEXO','Trabajo']).size().unstack())
agru_sex.plot(kind="bar", stacked=False)
plt.title("EGRESADOS COORDINADORES: EGRESADOS VS GENERO")
plt.xlabel("Genero")
plt.ylabel("Cantidad de Egresados")
plt.grid()
```
